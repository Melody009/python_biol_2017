```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_excel('http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.xls')
titanic.head()
```

```python
titanic_first_name = titanic['first_name'] = titanic['name'].str.split(', ').str.get(1).str.split('.').str.get(1)
titanic['first_name']
```

```python
name_length = titanic_first_name.str.split().str.len()
```

```python
name_length
```

```python
name_length.value_counts().reset_index()
```
